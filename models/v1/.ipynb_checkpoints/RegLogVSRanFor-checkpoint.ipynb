```python
%%time
import pandas as pd, numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, average_precision_score, precision_recall_curve, roc_curve
)
import matplotlib.pyplot as plt

# ==================== 1) CARGA ====================
df = pd.read_csv("../datasets/final/ico_union_canonical_v3.csv")
df = df[df["ico_successful"].notna()].copy()
df["ico_successful"] = df["ico_successful"].astype(int)

# (Re)calcular pct_goal_reached si faltara
if "pct_goal_reached" not in df.columns:
    df["pct_goal_reached"] = np.nan
if {"goal_usd","amount_raised_usd"}.issubset(df.columns):
    g = pd.to_numeric(df["goal_usd"], errors="coerce")
    a = pd.to_numeric(df["amount_raised_usd"], errors="coerce")
    with np.errstate(divide="ignore", invalid="ignore"):
        pct = a / g
    pct = pd.Series(pct).replace([np.inf, -np.inf], np.nan)
    df.loc[pct.notna(), "pct_goal_reached"] = pct

# ==================== 2) FEATURES ====================
bin_cols = ["has_github", "whitepaper_available"]
num_cols_base = ["pct_goal_reached","amount_raised_usd","goal_usd","team_size","token_price_usd"]
# (podés agregar más numéricas aquí si ya las tenés limpias: total_tokens, hard_cap_usd, etc.)

for c in num_cols_base:
    if c not in df.columns: df[c] = np.nan
for c in bin_cols:
    if c not in df.columns: df[c] = np.nan

def binarize(series):
    s = series.astype(str).str.strip().str.lower()
    mapping = {
        "true":1, "false":0, "yes":1, "no":0, "y":1, "n":0,
        "si":1, "sí":1, "present":1, "available":1,
        "absent":0, "unavailable":0, "none":np.nan, "nan":np.nan, "":np.nan
    }
    s = s.replace(mapping)
    s_num = pd.to_numeric(s, errors="coerce")
    out = np.where(s_num.notna(), (s_num > 0).astype(int), np.nan)
    return pd.Series(out, index=series.index, dtype="float")

# Convertir
for c in num_cols_base:
    df[c] = pd.to_numeric(df[c], errors="coerce").replace([np.inf, -np.inf], np.nan)
for c in bin_cols:
    df[c] = binarize(df[c]).fillna(0).astype(int)

# X, y
features_all = num_cols_base + bin_cols
X = df[features_all].copy()
y = df["ico_successful"].copy()

# Quitar columnas completamente vacías (si las hubiera)
all_nan_cols = X.columns[X.isna().all()].tolist()
if all_nan_cols:
    print("Cols all-NaN (drop):", all_nan_cols)
    X = X.drop(columns=all_nan_cols)

# Split consistente para ambos modelos
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# ==================== 3) PREP PRO LOGÍSTICA ====================
# Elegimos columnas a log-transform (muy sesgadas): montos/precios
log_cols = [c for c in ["amount_raised_usd","goal_usd","token_price_usd"] if c in X_train.columns]

def log1p_df(df_in, cols):
    df_out = df_in.copy()
    for c in cols:
        df_out[c] = pd.to_numeric(df_out[c], errors="coerce")
        df_out[c] = np.log1p(df_out[c])  # log(1+x)
    return df_out

# Imputación (mediana) y escalado (solo para LogReg)
imp = SimpleImputer(strategy="median")
scaler = StandardScaler()

Xtr_log = log1p_df(X_train, log_cols)
Xte_log = log1p_df(X_test,  log_cols)

Xtr_imp = imp.fit_transform(Xtr_log)
Xte_imp = imp.transform(Xte_log)

Xtr_sc  = scaler.fit_transform(Xtr_imp)
Xte_sc  = scaler.transform(Xte_imp)

# ==================== 4) MODELO A — LOGISTIC (balanced) ====================
logreg = LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42)
logreg.fit(Xtr_sc, y_train)
yp_log = logreg.predict(Xte_sc)
ys_log = logreg.predict_proba(Xte_sc)[:,1]

# ==================== 5) PREP PRO RANDOM FOREST ====================
# Para RF: imputación, sin escalar ni log (aunque podrías probar log también)
imp_rf = SimpleImputer(strategy="median")
Xtr_rf = imp_rf.fit_transform(X_train)
Xte_rf = imp_rf.transform(X_test)

# ==================== 6) MODELO B — RANDOM FOREST ====================
rf = RandomForestClassifier(
    n_estimators=300, max_depth=None, random_state=42,
    class_weight="balanced_subsample", min_samples_leaf=2
)
rf.fit(Xtr_rf, y_train)
yp_rf = rf.predict(Xte_rf)
ys_rf = rf.predict_proba(Xte_rf)[:,1]

# ==================== 7) MÉTRICAS COMPARADAS ====================
def report(name, y_true, y_pred, y_score):
    acc  = accuracy_score(y_true, y_pred)
    roc  = roc_auc_score(y_true, y_score)
    ap   = average_precision_score(y_true, y_score)  # PR-AUC
    print(f"\n{name}")
    print(f"  Accuracy: {acc:.3f} | ROC-AUC: {roc:.3f} | PR-AUC: {ap:.3f}")
    print("  Classification report:")
    print(classification_report(y_true, y_pred, digits=3))
    print("  Confusion Matrix:")
    print(confusion_matrix(y_true, y_pred))

report("Modelo A — Logistic (balanced + log1p + scaled)", y_test, yp_log, ys_log)
report("Modelo B — RandomForest (balanced_subsample)",   y_test, yp_rf,  ys_rf)

# ==================== 8) CURVAS ROC (opcional visual) ====================
fpr_log, tpr_log, _ = roc_curve(y_test, ys_log)
fpr_rf,  tpr_rf,  _ = roc_curve(y_test, ys_rf)
plt.figure(figsize=(6,5))
plt.plot(fpr_log, tpr_log, label="LogReg")
plt.plot(fpr_rf,  tpr_rf,  label="RandomForest")
plt.plot([0,1],[0,1],'k--',alpha=0.5)
plt.xlabel("FPR"); plt.ylabel("TPR"); plt.title("ROC curves"); plt.legend(); plt.show()

# ==================== 9) IMPORTANCIAS ====================
# Coefs de la logística (en espacio escalado+log): señal y dirección
coef_df = pd.DataFrame({"feature": X.columns, "coef": logreg.coef_[0]}).sort_values("coef", ascending=False)
print("\nCoeficientes LogReg (positivos ↑ prob. de éxito):")
print(coef_df.to_string(index=False))

# Importancias de RF
imp_rf = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nTop 15 Importancias RF:")
print(imp_rf.head(15).to_string())

plt.figure(figsize=(8,6))
imp_rf.head(15).plot(kind="barh")
plt.gca().invert_yaxis()
plt.title("Top 15 Feature Importances — RandomForest")
plt.show()
```

```python
import joblib
from pathlib import Path

bundle = joblib.load("../datasets/artifacts/ico_preproc_bundle.pk1")

X_train_prep = bundle.X_train_prep   # numpy array listo para modelar
X_test_prep  = bundle.X_test_prep
y_train      = bundle.y_train
y_test       = bundle.y_test

feat_names   = bundle.feature_names   # nombres alineados a X_*_prep
preproc      = bundle.preprocessor    # ColumnTransformer ya fit
cat_cols     = bundle.categorical_cols
num_cols     = bundle.numeric_cols
bin_cols     = bundle.binary_cols

# Reconstituir preprocesador con las mismas listas
def build_preprocessor(cats, nums, bins):
    cat_pipe = Pipeline([
        ("imp", SimpleImputer(strategy="most_frequent")),
        ("ohe", OneHotEncoder(handle_unknown="ignore", sparse_output=False)),
    ])
    num_pipe = Pipeline([
        ("imp", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler()),
    ])
    bin_pipe = Pipeline([
        ("imp", SimpleImputer(strategy="constant", fill_value=0)),
    ])
    transformers = []
    if nums: transformers.append(("num", num_pipe, nums))
    if bins: transformers.append(("bin", bin_pipe, bins))
    if cats: transformers.append(("cat", cat_pipe, cats))
    return ColumnTransformer(transformers, remainder="drop")
```

```python
%%time

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, average_precision_score
from sklearn.ensemble import RandomForestClassifier

pre = build_preprocessor(bundle.categorical_cols, bundle.numeric_cols, bundle.binary_cols)

rf = RandomForestClassifier(
    n_estimators=500,
    max_depth=None,
    min_samples_leaf=2,
    class_weight="balanced_subsample",
    random_state=42,
    n_jobs=-1
)

pipe_rf = Pipeline([
    ("pre", pre),
    ("clf", rf)
])

param_grid_rf = {
    "clf__n_estimators": [300, 600],
    "clf__max_depth": [None, 10, 20],
    "clf__min_samples_leaf": [1, 3, 5],
    "clf__max_features": ["sqrt", 0.5],
}

gs_rf = GridSearchCV(
    pipe_rf, param_grid_rf, scoring="roc_auc", cv=5, n_jobs=-1, verbose=0
)
gs_rf.fit(bundle.X_train, bundle.y_train)

y_pred_rf = gs_rf.predict(bundle.X_test)
y_proba_rf = gs_rf.predict_proba(bundle.X_test)[:,1]

acc_rf = accuracy_score(y_test, y_pred_rf)
roc_auc_rf = roc_auc_score(y_test, y_proba_rf)
pr_auc_rf = average_precision_score(y_test, y_proba_rf)

print("=== RandomForest (balanced_subsample) ===")
print(f"Accuracy: {acc_rf:.3f}  |  ROC-AUC: {roc_auc_rf:.3f}  |  PR-AUC: {pr_auc_rf:.3f}\n")
print("Classification Report:")
print(classification_report(y_test, y_pred_rf, digits=3))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
print("\nBest params LR:", gs_rf.best_params_)
```

```python
%%time

import pandas as pd, numpy as np
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, average_precision_score

pre = build_preprocessor(bundle.categorical_cols, bundle.numeric_cols, bundle.binary_cols)

# Ejemplo 1: LogisticRegression con grid
pipe_lr = Pipeline([
    ("pre", pre),
    ("clf", LogisticRegression(max_iter=2000, class_weight="balanced", n_jobs=-1))
])

param_grid_lr = {
    "clf__C": [0.1, 1, 3],
    "clf__penalty": ["l2"],
    "clf__solver": ["lbfgs", "liblinear"],
}

gs_lr = GridSearchCV(
    pipe_lr, param_grid_lr, scoring="roc_auc", cv=5, n_jobs=-1, verbose=0
)
gs_lr.fit(bundle.X_train, bundle.y_train)

y_pred_lr = gs_lr.predict(bundle.X_test)
y_proba_lr = gs_lr.predict_proba(bundle.X_test)[:,1]

acc_lr = accuracy_score(y_test, y_pred_lr)
roc_auc_lr = roc_auc_score(y_test, y_proba_lr)
pr_auc_lr = average_precision_score(y_test, y_proba_lr)

print("=== Logistic Regression (balanced) con GridSearchCV ===")
print(f"Accuracy: {acc_lr:.3f}  |  ROC-AUC: {roc_auc_lr:.3f}  |  PR-AUC: {pr_auc_lr:.3f}\n")
print("Report de Clasificacion:")
print(classification_report(y_test, y_pred_lr, digits=3))
print("Matriz de confusión:")
print(confusion_matrix(y_test, y_pred_lr))
print("\nBest params LR:", gs_lr.best_params_)
```

```python
#  Gráficos de resultados de modelos

from sklearn.metrics import roc_curve, auc, precision_recall_curve, ConfusionMatrixDisplay
from sklearn.model_selection import learning_curve, validation_curve
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd, numpy as np

# --- 1️⃣ Comparar Accuracy, ROC-AUC y PR-AUC entre modelos ---
results = pd.DataFrame({
    "Modelo": ["Regresión Logística", "Random Forest"],
    "Accuracy": [acc_lr, acc_rf],
    "ROC-AUC": [roc_auc_lr, roc_auc_rf],
    "PR-AUC": [pr_auc_lr, pr_auc_rf]
})

results_melted = results.melt(id_vars="Modelo", var_name="Métrica", value_name="Valor")
plt.figure(figsize=(7,5))
sns.barplot(data=results_melted, x="Métrica", y="Valor", hue="Modelo", palette="crest")
plt.title("Comparación de métricas entre modelos base")
plt.ylim(0.5, 0.9)
plt.show()

# --- Curva de aprendizaje (Learning Curve) ---
best_rf = gs_rf.best_estimator_.named_steps["clf"]

train_sizes, train_scores, test_scores = learning_curve(
    best_rf,
    bundle.X_train_prep, bundle.y_train,
    train_sizes=np.linspace(0.1, 1.0, 6),
    cv=5, scoring='roc_auc', n_jobs=-1
)

plt.figure(figsize=(7,5))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label="Entrenamiento")
plt.plot(train_sizes, np.mean(test_scores, axis=1), 'o-', label="Validación")
plt.title("Curva de aprendizaje - Random Forest")
plt.xlabel("Porcentaje de datos de entrenamiento")
plt.ylabel("ROC-AUC promedio")
plt.legend()
plt.show()

# --- Curva de validacion ---
param_range = [50, 100, 200, 400, 600, 800]
train_scores, test_scores = validation_curve(
    RandomForestClassifier(class_weight="balanced_subsample", random_state=42, n_jobs=-1),
    bundle.X_train_prep, bundle.y_train,
    param_name="n_estimators",
    param_range=param_range,
    scoring="roc_auc", cv=5
)

plt.figure(figsize=(7,5))
plt.plot(param_range, np.mean(train_scores, axis=1), label="Entrenamiento")
plt.plot(param_range, np.mean(test_scores, axis=1), label="Validación")
plt.title("Curva de validación - n_estimators (Random Forest)")
plt.xlabel("Número de árboles")
plt.ylabel("ROC-AUC promedio")
plt.legend()
plt.show()

# --- Matriz de correlación de errores entre modelos ---
preds_df = pd.DataFrame({
    "y_real": bundle.y_test,
    "pred_lr": y_pred_lr,
    "pred_rf": y_pred_rf
})
sns.heatmap(preds_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlación de errores entre modelos")
plt.show()

# --- 1️⃣ Curvas ROC ---
plt.figure(figsize=(7,6))
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_proba_lr)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_proba_rf)
plt.plot(fpr_lr, tpr_lr, label=f"Logistic Regression (AUC={auc(fpr_lr, tpr_lr):.3f})")
plt.plot(fpr_rf, tpr_rf, label=f"Random Forest (AUC={auc(fpr_rf, tpr_rf):.3f})")
plt.plot([0,1],[0,1],'--',color='gray')
plt.xlabel("Tasa de Falsos Positivos")
plt.ylabel("Tasa de Verdaderos Positivos")
plt.title("Curva ROC comparativa")
plt.legend()
plt.show()

# --- 2️⃣ Curvas Precision-Recall ---
plt.figure(figsize=(7,6))
prec_lr, rec_lr, _ = precision_recall_curve(y_test, y_proba_lr)
prec_rf, rec_rf, _ = precision_recall_curve(y_test, y_proba_rf)
plt.plot(rec_lr, prec_lr, label=f"Logistic Regression (PR-AUC={auc(rec_lr, prec_lr):.3f})")
plt.plot(rec_rf, prec_rf, label=f"Random Forest (PR-AUC={auc(rec_rf, prec_rf):.3f})")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Curva Precision-Recall comparativa")
plt.legend()
plt.show()

# --- 3️⃣ Matriz de confusión (Random Forest) ---
disp = ConfusionMatrixDisplay.from_predictions(y_test, y_pred_rf, cmap="Blues", normalize=None)
disp.ax_.set_title("Matriz de confusión - Random Forest")
plt.show()

# --- 4️⃣ Importancia de características ---
importances = gs_rf.best_estimator_.named_steps["clf"].feature_importances_
feat_names = gs_rf.best_estimator_.named_steps["pre"].get_feature_names_out()
indices = np.argsort(importances)[-15:]
plt.figure(figsize=(8,6))
plt.barh(np.array(bundle.feature_names)[indices], importances[indices], color='teal')
plt.title("Top 15 características más importantes (Random Forest)")
plt.xlabel("Importancia")
plt.show()
```
