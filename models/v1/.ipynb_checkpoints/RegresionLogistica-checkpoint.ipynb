```python
%%time
import pandas as pd, numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt

# ---------- 1) CARGA ----------
df = pd.read_csv("../datasets/final/ico_union_canonical_v3.csv")
df = df[df["ico_successful"].notna()].copy()
df["ico_successful"] = df["ico_successful"].astype(int)

# ---------- 2) (RE)DERIVADAS MÃNIMAS ----------
# Recalcular pct_goal_reached por si no estaba o estaba todo NaN
if "pct_goal_reached" not in df.columns:
    df["pct_goal_reached"] = np.nan
if "goal_usd" in df.columns and "amount_raised_usd" in df.columns:
    g = pd.to_numeric(df["goal_usd"], errors="coerce")
    a = pd.to_numeric(df["amount_raised_usd"], errors="coerce")
    with np.errstate(divide="ignore", invalid="ignore"):
        pct = a / g
    pct = pct.replace([np.inf, -np.inf], np.nan)
    # SÃ³lo sobreescribe donde se pueda calcular
    df.loc[pct.notna(), "pct_goal_reached"] = pct

# ---------- 3) FEATURES ----------
features = [
    "pct_goal_reached", "amount_raised_usd", "goal_usd",
    "team_size", "token_price_usd",
    "has_github", "whitepaper_available"
]
for f in features:
    if f not in df.columns:
        df[f] = np.nan

# binarize robusto
def binarize(series):
    s = series.astype(str).str.strip().str.lower()
    mapping = {
        "true":1, "false":0, "yes":1, "no":0, "y":1, "n":0,
        "si":1, "sÃ­":1, "present":1, "available":1,
        "absent":0, "unavailable":0, "none":np.nan, "nan":np.nan, "":np.nan
    }
    s = s.replace(mapping)
    s_num = pd.to_numeric(s, errors="coerce")
    out = np.where(s_num.notna(), (s_num > 0).astype(int), np.nan)
    return pd.Series(out, index=series.index, dtype="float")

num_cols = ["pct_goal_reached","amount_raised_usd","goal_usd","team_size","token_price_usd"]
bin_cols = ["has_github","whitepaper_available"]

# convertir numÃ©ricas
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce").replace([np.inf, -np.inf], np.nan)

# binarios â†’ 0/1 (default 0 si faltan)
for c in bin_cols:
    df[c] = binarize(df[c])
    df[c] = df[c].fillna(0).astype(int)

X = df[features].copy()
y = df["ico_successful"].copy()

# ---------- 4) DROP de columnas all-NaN ANTES de imputar ----------
all_nan_cols = X.columns[X.isna().all()].tolist()
if all_nan_cols:
    print("Cols all-NaN (dropeadas):", all_nan_cols)
    X = X.drop(columns=all_nan_cols)

# guardamos los nombres REALES que llegan al modelo
feature_names = X.columns.tolist()

# ---------- 5) SPLIT ----------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# ---------- 6) IMPUTACIÃ“N + ESCALA ----------
imp = SimpleImputer(strategy="median")
X_train_imp = imp.fit_transform(X_train)
X_test_imp  = imp.transform(X_test)

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train_imp)
X_test_sc  = scaler.transform(X_test_imp)

# ---------- 7) MODELO ----------
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train_sc, y_train)

# ---------- 8) EVAL ----------
y_pred = model.predict(X_test_sc)
acc = accuracy_score(y_test, y_pred)
print(f"\nðŸŽ¯ Accuracy: {acc:.3f}\n")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# ---------- 9) COEFICIENTES (alineados a feature_names) ----------
coef = model.coef_[0]
coef_df = pd.DataFrame({"feature": feature_names, "coef": coef}).sort_values("coef", ascending=False)
print("\nCoeficientes (positivos = â†‘ prob. de Ã©xito):")
print(coef_df.to_string(index=False))

plt.figure(figsize=(8,5))
colors = ["green" if c>0 else "red" for c in coef_df["coef"]]
plt.barh(coef_df["feature"], coef_df["coef"], color=colors)
plt.title("Pesos del modelo logÃ­stico")
plt.gca().invert_yaxis()
plt.show()
```

```python
%%time
# --- LOGISTIC REGRESSION (EX-ANTE) ---
import pandas as pd, numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, average_precision_score
)

# 1) Cargar dataset EX-ANTE (no incluye columnas leaky)
# DF_PATH = "../datasets/final/ico_exante_features_v1.csv"
DF_PATH = "../datasets/final/ico_dataset_preico_features.csv"
TARGET  = "ico_successful"
ID_COLS = ["name_std","symbol_std", TARGET]

df = pd.read_csv(DF_PATH)
df = df[df['ico_successful'].notna()]

# 2) Derivar features calendario a partir de la fecha de inicio (si existe)
drop_date_cols = []
if "ico_start_date" in df.columns:
    df["ico_start_date"] = pd.to_datetime(df["ico_start_date"], errors="coerce", dayfirst=True)
    df["start_year"]  = df["ico_start_date"].dt.year
    df["start_month"] = df["ico_start_date"].dt.month
    df["start_qtr"]   = df["ico_start_date"].dt.quarter
    df["start_dow"]   = df["ico_start_date"].dt.weekday
    drop_date_cols.append("ico_start_date")  # no usamos la fecha cruda

# 3) Separar X / y
df = df.copy()
y = df[TARGET].astype("Int64").astype(float).astype(int)
X = df.drop(columns=[c for c in ID_COLS if c in df.columns] + drop_date_cols, errors="ignore")

# 4) Detectar num/cat
num_cols = X.select_dtypes(include=["number"]).columns.tolist()
cat_cols = X.select_dtypes(exclude=["number"]).columns.tolist()

# 5) Preprocesamiento + modelo
num_pipe = Pipeline(steps=[
    ("imp", SimpleImputer(strategy="median")),
    ("sc",  StandardScaler())
])
cat_pipe = Pipeline(steps=[
    ("imp", SimpleImputer(strategy="most_frequent")),
    ("oh",  OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])
prep = ColumnTransformer(
    transformers=[("num", num_pipe, num_cols),
                  ("cat", cat_pipe, cat_cols)],
    remainder="drop"
)

clf = LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42)

pipe = Pipeline(steps=[("prep", prep), ("clf", clf)])

# 6) Split estratificado (este notebook es auto-contenido)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# 7) Fit + evaluaciÃ³n
pipe.fit(X_train, y_train)
y_pred  = pipe.predict(X_test)
y_proba = pipe.predict_proba(X_test)[:, 1]

print("=== Logistic Regression (balanced) ===")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.3f}  |  ROC-AUC: {roc_auc_score(y_test, y_proba):.3f}  |  PR-AUC: {average_precision_score(y_test, y_proba):.3f}\n")
print("Classification Report:")
print(classification_report(y_test, y_pred, digits=3))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# 8) Coeficientes con nombres *robustos* (del ColumnTransformer ya entrenado)
#    Esto evita el desfasaje que te tiraba el ValueError.
feature_names = pipe.named_steps["prep"].get_feature_names_out()  # ej: "num__goal_usd", "cat__jurisdiction_<cat>"
coefs = pipe.named_steps["clf"].coef_[0]
coef_df = pd.DataFrame({"feature": feature_names, "coef": coefs})
coef_df["abs"] = coef_df["coef"].abs()
coef_df = coef_df.sort_values("abs", ascending=False).drop(columns="abs")

print("\nTop 20 coeficientes (Â±):")
print(coef_df.head(20).to_string(index=False))

print(f"Categorical columnas usadas: {cat_cols}. ")
print(f"Numerical columnas usadas: {num_cols}. ")
print(f'Distribucion de ico_successful: {df["ico_successful"].value_counts(dropna=False)}')
```

```python
import joblib
from pathlib import Path

bundle = joblib.load("../datasets/artifacts/ico_preproc_bundle.pk1")

X_train_prep = bundle.X_train_prep   # numpy array listo para modelar
X_test_prep  = bundle.X_test_prep
y_train      = bundle.y_train
y_test       = bundle.y_test

feat_names   = bundle.feature_names   # nombres alineados a X_*_prep
preproc      = bundle.preprocessor    # ColumnTransformer ya fit
cat_cols     = bundle.categorical_cols
num_cols     = bundle.numeric_cols
bin_cols     = bundle.binary_cols

# Reconstituir preprocesador con las mismas listas
def build_preprocessor(cats, nums, bins):
    cat_pipe = Pipeline([
        ("imp", SimpleImputer(strategy="most_frequent")),
        ("ohe", OneHotEncoder(handle_unknown="ignore", sparse_output=False)),
    ])
    num_pipe = Pipeline([
        ("imp", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler()),
    ])
    bin_pipe = Pipeline([
        ("imp", SimpleImputer(strategy="constant", fill_value=0)),
    ])
    transformers = []
    if nums: transformers.append(("num", num_pipe, nums))
    if bins: transformers.append(("bin", bin_pipe, bins))
    if cats: transformers.append(("cat", cat_pipe, cats))
    return ColumnTransformer(transformers, remainder="drop")
```

```python
%%time

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, average_precision_score

logreg = LogisticRegression(max_iter=2000, class_weight="balanced", n_jobs=-1)
logreg.fit(X_train_prep, y_train)

y_pred = logreg.predict(X_test_prep)
y_proba = logreg.predict_proba(X_test_prep)[:,1]

print("=== Logistic Regression (balanced) ===")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.3f}  |  ROC-AUC: {roc_auc_score(y_test, y_proba):.3f}  |  PR-AUC: {average_precision_score(y_test, y_proba):.3f}\n")
print("Report de Clasificacion:")
print(classification_report(y_test, y_pred, digits=3))
print("Matriz de confusiÃ³n:")
print(confusion_matrix(y_test, y_pred))
```

```python
%%time

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, average_precision_score

pre = build_preprocessor(bundle.categorical_cols, bundle.numeric_cols, bundle.binary_cols)

# Ejemplo 1: LogisticRegression con grid
pipe_lr = Pipeline([
    ("pre", pre),
    ("clf", LogisticRegression(max_iter=2000, class_weight="balanced", n_jobs=-1))
])

param_grid_lr = {
    "clf__C": [0.1, 1, 3],
    "clf__penalty": ["l2"],
    "clf__solver": ["lbfgs", "liblinear"],
}

gs_lr = GridSearchCV(
    pipe_lr, param_grid_lr, scoring="roc_auc", cv=5, n_jobs=-1, verbose=0
)
gs_lr.fit(bundle.X_train, bundle.y_train)

y_pred = gs_lr.predict(X_test_prep)
y_proba = gs_lr.predict_proba(X_test_prep)[:,1]

print("=== Logistic Regression (balanced) con GridSearchCV ===")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.3f}  |  ROC-AUC: {roc_auc_score(y_test, y_proba):.3f}  |  PR-AUC: {average_precision_score(y_test, y_proba):.3f}\n")
print("Report de Clasificacion:")
print(classification_report(y_test, y_pred, digits=3))
print("Matriz de confusiÃ³n:")
print(confusion_matrix(y_test, y_pred))
print("\nBest params LR:", gs_lr.best_params_)
```
