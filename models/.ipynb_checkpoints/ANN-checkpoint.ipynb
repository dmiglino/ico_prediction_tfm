```python
import os, json, time, joblib
import numpy as np, pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_curve, roc_auc_score, precision_recall_curve, average_precision_score,
    f1_score, brier_score_loss
)
from sklearn.calibration import CalibratedClassifierCV, calibration_curve

# ---------- Funcion para imprimir resultados por pantalla ------------ 
def print_metrics(y_true, y_pred, y_proba=None, title=None):
    if title: print(f"\n=== {title} ===")
    acc = accuracy_score(y_true, y_pred)
    print(f"\nðŸŽ¯ Accuracy: {acc:.3f}")
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, digits=4))
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_true, y_pred))
    if y_proba is not None:
        print("\nProb-metrics:")
        print(f"ROC-AUC : {roc_auc_score(y_true, y_proba):.4f}")
        print(f"PR-AUC  : {average_precision_score(y_true, y_proba):.4f}")
        print(f"Brier   : {brier_score_loss(y_true, y_proba):.4f}")

def save_experiment(exp_dir, model_tag, best_estimator, y_true, y_pred, y_proba, metrics_extra=None):
    os.makedirs(exp_dir, exist_ok=True)
    mets = {
        "timestamp": time.strftime("%Y-%m-%d %H:%M:%S"),
        "model": model_tag,
        "accuracy": float(accuracy_score(y_true, y_pred)),
        "roc_auc": float(roc_auc_score(y_true, y_proba)),
        "pr_auc": float(average_precision_score(y_true, y_proba)),
        "f1": float(f1_score(y_test, y_pred)),
        "brier": float(brier_score_loss(y_true, y_proba)),
        "exp_dir": exp_dir,
    }
    if metrics_extra: mets.update(metrics_extra)
    pd.DataFrame({"y_true": y_true, "y_pred": y_pred, "y_proba": y_proba}).to_csv(
        os.path.join(exp_dir, "test_predictions.csv"), index=False
    )
    with open(os.path.join(exp_dir, "metrics.json"), "w") as f:
        json.dump({model_tag: mets}, f, indent=2)
    joblib.dump(best_estimator, os.path.join(exp_dir, f"{model_tag}.pkl"))

    os.makedirs("experiments", exist_ok=True)
    lb_path = "experiments/leaderboard.csv"
    row = {k: mets[k] for k in ["timestamp","model","accuracy","roc_auc","pr_auc","brier","exp_dir"]}
    df = pd.DataFrame([row])
    if os.path.exists(lb_path):
        df.to_csv(lb_path, mode="a", header=False, index=False)
    else:
        df.to_csv(lb_path, index=False)
    return mets
```

# ANN (MLPClassifier): Grid + CalibraciÃ³n + Save #
# ============================================== #

```python
%%time

# ---------- Paths
DATA_PATH = "../datasets/final/ico_dataset_final_v2_clean_enriquecido_feature_engineering_preico_v1.csv"
STAMP = time.strftime("%Y%m%d_%H%M%S")
EXP_DIR = f"experiments/ann_v1"
FIG_DIR = os.path.join(EXP_DIR, "figs")
os.makedirs(FIG_DIR, exist_ok=True)

# ---------- Load
df = pd.read_csv(DATA_PATH)
target = "ico_successful"
df[target] = df[target].astype(int)
df = df.dropna(subset=[target])

# ---------- Feature blocks (idÃ©ntico a los otros modelos)
cat_cols = [c for c in ["platform","category","location","caps_unit"] if c in df.columns]
bin_cols = [c for c in ["mvp","has_twitter","has_facebook","is_tax_regulated","has_github",
                        "has_reddit","has_website","has_whitepaper","kyc",
                        "accepts_BTC","accepts_ETH","has_contract_address"] if c in df.columns]
num_cols = [c for c in df.columns if c not in cat_cols + bin_cols + [target]]

# ---------- Preprocess (ANN necesita escalado en numÃ©ricas)
num_pipe = Pipeline([("imputer", SimpleImputer(strategy="median")), ("scaler", StandardScaler())])
cat_pipe = Pipeline([("imputer", SimpleImputer(strategy="most_frequent")),
                     ("ohe", OneHotEncoder(handle_unknown="ignore", drop="first"))])
bin_pipe = Pipeline([("imputer", SimpleImputer(strategy="constant", fill_value=0))])
pre = ColumnTransformer([("num", num_pipe, num_cols), ("cat", cat_pipe, cat_cols), ("bin", bin_pipe, bin_cols)],
                        remainder="drop", sparse_threshold=0.3)

X = df.drop(columns=[target]); y = df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, stratify=y, random_state=42)
print(f"Variables finales: {X.shape[1]}  |  Filas usadas: {X.shape[0]}")

# ---------- Modelo base (MLP)
mlp = MLPClassifier(
    max_iter=600,          # early_stopping corta antes si converge
    early_stopping=True,   # valida internamente
    n_iter_no_change=30,
    random_state=42,
)

mlp_pipe = Pipeline([("pre", pre), ("clf", mlp)])

# ---------- Grid de hiperparÃ¡metros (tocÃ¡ lo que quieras)
# hidden_layer_sizes define arquitectura: (neuronas_capa1, neuronas_capa2, ...)
param_grid = {
    "clf__hidden_layer_sizes": [
        (128, 64),
        (128, 64),
        (128, 64, 32),      # un poco mÃ¡s profunda
        (256, 128),         # mÃ¡s capacidad
    ],
    "clf__activation": ["relu", "tanh"],
    "clf__alpha": [1e-5, 1e-4, 1e-3],        # L2 (weight decay)
    "clf__learning_rate_init": [1e-3, 3e-3], # LR base
    "clf__batch_size": [64, 128, 256],
    "clf__solver": ["adam"],                 # (lbfgs no escala bien en OHE grandes)
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# *** Usamos PR-AUC como mÃ©trica de selecciÃ³n ***
gs = GridSearchCV(mlp_pipe, param_grid, cv=cv, scoring="average_precision", n_jobs=-1, refit=True)
gs.fit(X_train, y_train)

best_ann = gs.best_estimator_
print(f"[ANN] Best params: {gs.best_params_}")

# ---- Eval sin calibrar
y_proba = best_ann.predict_proba(X_test)[:, 1]
y_pred  = (y_proba >= 0.5).astype(int)
print_metrics(y_test, y_pred, y_proba, title="ANN (raw)")

# ---------- CalibraciÃ³n isotÃ³nica (mejora Brier/probs)
cal = CalibratedClassifierCV(estimator=best_ann.named_steps["clf"], method="isotonic", cv=5)
ann_cal = Pipeline([("pre", pre), ("cal", cal)])
ann_cal.fit(X_train, y_train)

y_proba_cal = ann_cal.predict_proba(X_test)[:, 1]
y_pred_cal  = (y_proba_cal >= 0.5).astype(int)
print_metrics(y_test, y_pred_cal, y_proba_cal, title="ANN (calibrated)")

# ---------- GrÃ¡ficos (ROC, PR, CalibraciÃ³n)
# ROC
fpr_raw, tpr_raw, _ = roc_curve(y_test, y_proba)
fpr_cal, tpr_cal, _ = roc_curve(y_test, y_proba_cal)
plt.figure()
plt.plot(fpr_raw, tpr_raw, label=f"ANN (AUC={roc_auc_score(y_test, y_proba):.3f})")
plt.plot(fpr_cal, tpr_cal, label=f"ANN+Cal (AUC={roc_auc_score(y_test, y_proba_cal):.3f})")
plt.plot([0,1],[0,1], linestyle="--")
plt.xlabel("FPR"); plt.ylabel("TPR"); plt.title("ROC - ANN"); plt.legend()
plt.show()
plt.savefig(os.path.join(FIG_DIR, "roc_ann.png"), bbox_inches="tight"); plt.close()

# PR
p_raw, r_raw, _ = precision_recall_curve(y_test, y_proba)
p_cal, r_cal, _ = precision_recall_curve(y_test, y_proba_cal)
plt.figure()
plt.plot(r_raw, p_raw, label=f"ANN (AP={average_precision_score(y_test, y_proba):.3f})")
plt.plot(r_cal, p_cal, label=f"ANN+Cal (AP={average_precision_score(y_test, y_proba_cal):.3f})")
plt.xlabel("Recall"); plt.ylabel("Precision"); plt.title("PR - ANN"); plt.legend()
plt.show()
plt.savefig(os.path.join(FIG_DIR, "pr_ann.png"), bbox_inches="tight"); plt.close()

# Calibration curve
prob_true, prob_pred = calibration_curve(y_test, y_proba_cal, n_bins=10, strategy="uniform")
plt.figure()
plt.plot(prob_pred, prob_true, marker="o", linewidth=1, label="Calibrated")
plt.plot([0,1],[0,1], linestyle="--")
plt.xlabel("Predicted probability"); plt.ylabel("Observed frequency")
plt.title("Calibration - ANN (Isotonic)"); plt.legend()
plt.show()
plt.savefig(os.path.join(FIG_DIR, "calibration_ann.png"), bbox_inches="tight"); plt.close()

# ---------- Guardado de artefactos + leaderboard
# _ = save_experiment(
#     EXP_DIR, "ann_mlp", best_ann,
#     y_test.values, y_pred, y_proba,
#     metrics_extra={"best_params": gs.best_params_}
# )

# # ---------- Guardado de artefactos + leaderboard
# _ = save_experiment(
#     EXP_DIR, "ann_isotonic_calibrated", ann_cal,
#     y_test.values, y_pred_cal, y_proba_cal,
#     metrics_extra={"best_params": gs.best_params_}
# )


m_ann = {
    "model": "ann_mlp",
    "best_params": gs.best_params_,
    "roc_auc": float(roc_auc_score(y_test, y_proba)),
    "pr_auc": float(average_precision_score(y_test, y_proba)),
    "f1": float(f1_score(y_test, y_pred)),
    "accuracy": float(accuracy_score(y_test, y_pred)),
    "brier": float(brier_score_loss(y_test, y_proba)),
    "confusion_matrix": confusion_matrix(y_test, y_pred).tolist(),
}

m_ann_cal = {
    "model": "ann_calibrated_isotonic",
    "roc_auc": float(roc_auc_score(y_test, y_proba_cal)),
    "pr_auc": float(average_precision_score(y_test, y_proba_cal)),
    "f1": float(f1_score(y_test, y_pred_cal)),
    "accuracy": float(accuracy_score(y_test, y_pred_cal)),
    "brier": float(brier_score_loss(y_test, y_proba_cal)),
    "confusion_matrix": confusion_matrix(y_test, y_pred_cal).tolist(),
}

os.makedirs("experiments", exist_ok=True)
lb_path = "experiments/leaderboard.csv"
row = {k: m_ann[k] for k in ["model","accuracy","roc_auc","pr_auc","f1","brier"]}
df = pd.DataFrame([row])
if os.path.exists(lb_path):
    df.to_csv(lb_path, mode="a", header=False, index=False)
else:
    df.to_csv(lb_path, index=False)
    
os.makedirs("experiments", exist_ok=True)
lb_path = "experiments/leaderboard.csv"
row = {k: m_ann_cal[k] for k in ["model","accuracy","roc_auc","pr_auc","f1","brier"]}
df = pd.DataFrame([row])
if os.path.exists(lb_path):
    df.to_csv(lb_path, mode="a", header=False, index=False)
else:
    df.to_csv(lb_path, index=False)

# ---------- Guardar artefactos
os.makedirs(EXP_DIR, exist_ok=True)
pd.DataFrame({
    "y_true": y_test.values,
    "proba_ann": y_proba,
    "proba_ann_cal": y_proba_cal
}).to_csv(os.path.join(EXP_DIR, "test_predictions.csv"), index=False)

with open(os.path.join(EXP_DIR, "metrics.json"), "w") as f:
    json.dump({"ann_mlp": m_ann, "ann_calibrated": m_ann_cal}, f, indent=2)

joblib.dump(gs.best_estimator_, os.path.join(EXP_DIR, "best_rf.pkl"))
joblib.dump(ann_cal, os.path.join(EXP_DIR, "rf_isotonic_calibrated.pkl"))

print("\nArtefactos guardados en:", EXP_DIR)
```
