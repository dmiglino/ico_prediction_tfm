### Busqueda de duplicados (symbol presente en ambos datasets)

##### El dataset JOINED es el resultado de la union de los datasets originales: Zenodo + Villanueva + Yun
##### El dataset FOUNDICO es el catalogo detallado y preprocesado de la API de FoundICO

```python
import pandas as pd
import random

# === Paths ===
JOINED_PATH = "../join/ico_union_canonical_v6.csv" # tiene features no pre-ICO como 'amount_raised', 'hit_hard_cap', 'hit_soft_cap' 
FOUNDICO_PATH = "../foundico/foundico_details.csv"

# === Cargar datasets ===
joined = pd.read_csv(JOINED_PATH)
foundi = pd.read_csv(FOUNDICO_PATH)

# Normalizar nombres
joined.columns = joined.columns.str.strip().str.lower()
foundi.columns = foundi.columns.str.strip().str.lower()

# Normalizar symbol-ticker
joined['symbol_std'] = joined['symbol_std'].astype(str).str.upper().str.strip()
foundi['symbol'] = foundi['symbol'].astype(str).str.upper().str.strip()
foundi.rename(columns={'symbol': 'symbol_std'}, inplace=True)

# Duplicados
duplis = sorted(set(joined['symbol_std']).intersection(set(foundi['symbol_std'])))
print(f"üîé ICO duplicadas en ambos datasets (FoundICO y la union de los demas): {len(duplis)}")

# Guardar lista para siguientes celdas
duplis_list = list(duplis)
print(random.sample(duplis_list, 40))  # muestra 40 aleatorios
```

### Enriquecer el dataset de FoundICO con info del dataset joineado (para los tokens que estan en ambos)

##### Se buscaran: 'ico_successfull', 'soft_cap', 'hard_cap', 'amount_raised', 'hit_hard_cap', 'hit_soft_cap'

```python
def imputar_valores_ico_successful(df, min_score=3.5, max_score=7.5):
    ### Si tiene datos en la columna markets significa que fue listado y por lo tanto tuvo exito ###
    # print(df.columns.tolist())
    
    # Candidatos: tienen hard_cap y raised, y a√∫n no tienen etiqueta
    con_markets = df["ico_successful"].isna() & df["markets"].notna()
    
    # Trazabilidad del origen de la imputaci√≥n
    df.loc[con_markets, "ico_successful_rule"] = 'markets'
    
    # Actualizar s√≥lo esos casos a √©xito (1)
    df.loc[con_markets, "ico_successful"] = 1
    #display(df[con_markets].loc[:,['raised','soft_cap','hard_cap','symbol_std','markets','ico_successful','hit_soft_cap','hit_hard_cap']])
    print(f"‚úÖ Etiquetados como √©xito por markets: {int(con_markets.sum())} ICOs.")
    
    
    ### Para los que tengan soft_cap y raised, pero no tengan valor en ico_successful, se puede calcular ###
    
    # Filtro de ICOs con soft_cap y raised presentes, pero sin ico_successful
    raised_name = 'raised' if 'raised' in df.columns else 'amount_raised'
    print(f"raised_name: {raised_name}")
    mask = df["soft_cap"].notna() & df[raised_name].notna() & df["ico_successful"].isna()
    
    # Calcular nuevo valor seg√∫n condici√≥n
    df.loc[mask, "ico_successful"] = np.where(
        df.loc[mask, raised_name] >= df.loc[mask, "soft_cap"], 
        1,  # √©xito
        0   # fracaso
    )
    
    # Trazabilidad del origen de la imputaci√≥n
    df.loc[mask, "ico_successful_rule"] = 'soft_cap'
    
    # Confirmar cambios en ico_successful
    n_set = mask.sum()
    print(f"‚úÖ Etiquetados como √©xito o fracaso por soft_cap: {n_set} ICOs.")
    
    
    ### En el caso de hard_cap y raised, solo se puede asegurar el exito cuando raised es mayor, pero si es menor no se puede asumir fracaso ###
    
    # Candidatos: tienen hard_cap y raised, y a√∫n no tienen etiqueta
    cand = df["ico_successful"].isna() & df["hard_cap"].notna() & df[raised_name].notna()
    
    # √âxito seguro por hard cap: raised >= hard_cap
    success_hardcap = cand & (df[raised_name] >= df["hard_cap"])
    
    # Trazabilidad del origen de la imputaci√≥n
    df.loc[success_hardcap, "ico_successful_rule"] = 'hard_cap'
    
    # Actualizar s√≥lo esos casos a √©xito (1)
    df.loc[success_hardcap, "ico_successful"] = 1
    
    print(f"‚úÖ Etiquetados como √©xito por hard_cap: {int(success_hardcap.sum())} ICOs.")
    
    # Para auditar: cu√°ntos candidatos restantes NO se pudieron etiquetar
    remaining_cand = cand & ~success_hardcap
    print(f"‚ÑπÔ∏è Candidatos no etiquetados (no superan hard_cap, podr√≠an haber superado soft_cap): {int(remaining_cand.sum())}")
    
    
    ### Si el ico_score es mayor a 6, se considera que es una ICO exitosa. Igualmente, voy a usar un umbral mas alto de 7.5 para asegurar el exito ### 
    
    # Candidatos: tienen hard_cap y raised, y a√∫n no tienen etiqueta
    con_ico_score = df["ico_successful"].isna() & df["ico_score"].notna()
    
    # √âxito seguro por hard cap: raised >= hard_cap
    ico_score_alto = con_ico_score & (df["ico_score"] >= max_score)
    ico_score_bajo = con_ico_score & (df["ico_score"] < min_score)
    
    # Trazabilidad del origen de la imputaci√≥n
    df.loc[ico_score_alto, "ico_successful_rule"] = 'ico_score_alto'
    df.loc[ico_score_bajo, "ico_successful_rule"] = 'ico_score_bajo'
    
    # Actualizar s√≥lo esos casos a √©xito (1)
    df.loc[ico_score_alto, "ico_successful"] = 1
    df.loc[ico_score_bajo, "ico_successful"] = 0
    
    print(f"‚úÖ Etiquetados como √©xito por ico_score_alto: {int(ico_score_alto.sum())} ICOs.")
    print(f"‚úÖ Etiquetados como √©xito por ico_score_bajo: {int(ico_score_bajo.sum())} ICOs.")
    
    # Para auditar: cu√°ntos candidatos restantes NO se pudieron etiquetar
    ico_score_medio = con_ico_score & ~ico_score_alto & ~ico_score_bajo
    print(f"‚ÑπÔ∏è Candidatos no etiquetados por ico_score_medio: {int(ico_score_medio.sum())}")
```

```python
# === Enriquecer FoundICO con datos del dataset Joineado ===
import pandas as pd
import numpy as np

# --- helpers ---
def pick(df, candidates):
    for c in candidates:
        if c in df.columns:
            return c
    return None

def is_missing(x):
    if pd.isna(x): 
        return True
    s = str(x).strip().lower()
    return s in {"", "nan", "none", "null"}

# Para asegurarse de que ambos datasets tengan la columna symbol_std
if "symbol_std" not in joined.columns:
    raise ValueError("joined no tiene la columna symbol_std")
if "symbol_std" not in foundi.columns:
    raise ValueError("foundi no tiene la columna symbol_std")

# print(joined.columns.tolist())
# Chequeo de columnas de inter√©s en el dataset joined
if "ico_successful" not in joined.columns:
    raise ValueError("No existe la columna de 'ico_successful' en el dataset joineado.")
if "soft_cap" not in joined.columns:
    raise ValueError("No existe la columna de 'soft_cap' en el dataset joineado.")
if "hard_cap" not in joined.columns:
    raise ValueError("No existe la columna de 'hard_cap' en el dataset joineado.")
if "amount_raised" not in joined.columns:
    raise ValueError("No existe la columna de 'amount_raised' en el dataset joineado.")
if "hit_soft_cap" not in joined.columns:
    raise ValueError("No existe la columna de 'hit_soft_cap' en el dataset joineado.")
if "hit_hard_cap" not in joined.columns:
    raise ValueError("No existe la columna de 'hit_hard_cap' en el dataset joineado.")

# Armar diccionarios por s√≠mbolo
ico_success_map  = joined.set_index("symbol_std")["ico_successful"].to_dict()
soft_cap_map = joined.set_index("symbol_std")["soft_cap"].to_dict()
hard_cap_map = joined.set_index("symbol_std")["hard_cap"].to_dict()
amount_raised_map = joined.set_index("symbol_std")["amount_raised"].to_dict()
hit_soft_cap_map = joined.set_index("symbol_std")["hit_soft_cap"].to_dict()
hit_hard_cap_map = joined.set_index("symbol_std")["hit_hard_cap"].to_dict()

# Asegurar columna target en el dataset de FoundICO
if "ico_successful" not in foundi.columns:
    foundi["ico_successful"] = np.nan
    
# Asegurar columna amount_raised en el dataset de FoundICO
# if "amount_raised" not in foundi.columns:
#     foundi["amount_raised"] = np.nan
    
# Asegurar columna hit_soft_cap en el dataset de FoundICO
if "hit_soft_cap" not in foundi.columns:
    foundi["hit_soft_cap"] = np.nan
    
# Asegurar columna hit_hard_cap en el dataset de FoundICO
if "hit_hard_cap" not in foundi.columns:
    foundi["hit_hard_cap"] = np.nan

# Si no existe duplis_list lo calculamos
if "duplis_list" not in globals():
    duplis_list = sorted(set(joined["symbol_std"]).intersection(set(foundi["symbol_std"])))

# Filtros solo a duplicados
mask_dup = foundi["symbol_std"].isin(duplis_list)
sub = foundi.loc[mask_dup, ["symbol_std", "ico_successful", "soft_cap", "hard_cap", "raised", "hit_soft_cap", "hit_hard_cap"]].copy()
#display(sub)

# Contadores antes
before_ico_success_na   = sub["ico_successful"].isna().sum()
before_soft_cap_na  = sub["soft_cap"].apply(is_missing).sum()
before_hard_cap_na  = sub["hard_cap"].apply(is_missing).sum()
before_raised_na  = sub["raised"].apply(is_missing).sum()

# Rellenar columna ico_successful desde joined
def fill_ico_successful(symbol, curr_val):
    v = ico_success_map.get(symbol, np.nan)
    return v if not pd.isna(v) else curr_val

sub["ico_successful"] = [
    fill_ico_successful(sym, curr) for sym, curr in zip(sub["symbol_std"], sub["ico_successful"])
]

# Rellenar columna amount_raised desde joined
# def fill_amount_raised(symbol, curr_val):
#     v = amount_raised_map.get(symbol, np.nan)
#     return v if not pd.isna(v) else curr_val
    
# sub["amount_raised"] = [
#     fill_amount_raised(sym, curr) for sym, curr in zip(sub["symbol_std"], sub["amount_raised"])
# ]

# Rellenar columna hit_soft_cap desde joined
def fill_hit_soft_cap(symbol, curr_val):
    v = hit_soft_cap_map.get(symbol, np.nan)
    return v if not pd.isna(v) else curr_val
    
sub["hit_soft_cap"] = [
    fill_hit_soft_cap(sym, curr) for sym, curr in zip(sub["symbol_std"], sub["hit_soft_cap"])
]

# Rellenar columna hit_hard_cap desde joined
def fill_hit_hard_cap(symbol, curr_val):
    v = hit_hard_cap_map.get(symbol, np.nan)
    return v if not pd.isna(v) else curr_val
    
sub["hit_hard_cap"] = [
    fill_hit_hard_cap(sym, curr) for sym, curr in zip(sub["symbol_std"], sub["hit_hard_cap"])
]

# Rellenar columna amount_raised desde joined
def fill_raised(symbol, curr_val):
    if not is_missing(curr_val):
        return curr_val
    v = amount_raised_map.get(symbol, np.nan)
    if v < 100 and v > 0:
        v = v * 1000000
    return v
    
sub["raised"] = [
    fill_raised(sym, curr) for sym, curr in zip(sub["symbol_std"], sub["raised"])
]

# Rellenar softcap s√≥lo si est√° vac√≠o en foundico y existe en joined
def fill_soft_cap(sym, curr_val):
    if not is_missing(curr_val):
        return curr_val
    v = soft_cap_map.get(sym, np.nan)
    return v

sub["soft_cap"] = [
    fill_soft_cap(sym, curr) for sym, curr in zip(sub["symbol_std"], sub["soft_cap"])
]

# Rellenar hardcap s√≥lo si est√° vac√≠o en foundico y existe en joined
def fill_soft_hard_cap(sym, curr_val):
    if not is_missing(curr_val):
        return curr_val
    v = hard_cap_map.get(sym, np.nan)
    return v
    
sub["hard_cap"] = [
    fill_soft_hard_cap(sym, curr) for sym, curr in zip(sub["symbol_std"], sub["hard_cap"])
]

# Escribir de vuelta al dataframe original
foundi.loc[mask_dup, "ico_successful"] = sub["ico_successful"].values
foundi.loc[mask_dup, "soft_cap"] = sub["soft_cap"].values
foundi.loc[mask_dup, "hard_cap"] = sub["hard_cap"].values
foundi.loc[mask_dup, "raised"] = sub["raised"].values
foundi.loc[mask_dup, "hit_soft_cap"] = sub["hit_soft_cap"].values
foundi.loc[mask_dup, "hit_hard_cap"] = sub["hit_hard_cap"].values


# Se imputan valor del target cuando sea posible
imputar_valores_ico_successful(foundi)


### Se pueden derivar flags binarios segun presencia de enlaces y redes, siguiendo el formato del dataset joineado ### 

cols_links = ["website", "whitepaper", "github", "twitter", "facebook", "reddit"]

for col in cols_links:
    new_col = f"has_{col}"
    if col in foundi.columns:
        foundi[new_col] = (
            foundi[col]
            .astype(str)
            .str.strip()
            .replace({"nan": np.nan, "": np.nan})
            .notna()
            .astype("Int64")
        )
    else:
        # si falta la columna original, crear la nueva vac√≠a
        foundi[new_col] = pd.Series(pd.NA, index=foundi.index, dtype="Int64")


# Reporte
after_ico_success_na  = foundi.loc[mask_dup, "ico_successful"].isna().sum()
after_soft_cap_na  = sub["soft_cap"].apply(is_missing).sum()
after_hard_cap_na  = sub["hard_cap"].apply(is_missing).sum()
after_raised_na  = sub["raised"].apply(is_missing).sum()

print(f"üéØ Duplicados procesados: {mask_dup.sum()}")
print(f"ico_successful:  antes NaN={before_ico_success_na}  ‚Üí  despu√©s NaN={after_ico_success_na}")
print(f"soft_cap: antes vac√≠os={before_soft_cap_na}  ‚Üí  despu√©s vac√≠os={after_soft_cap_na}")
print(f"hard_cap: antes vac√≠os={before_hard_cap_na}  ‚Üí  despu√©s vac√≠os={after_hard_cap_na}")
print(f"raised: antes vac√≠os={before_raised_na}  ‚Üí  despu√©s vac√≠os={after_raised_na}")

# Renombre de columnas para consistencia con el dataset joineado
foundi.rename(columns={'raised': 'amount_raised', 
                       'name': 'name_std'
                      }, 
              inplace=True)


# Persistir
OUT_FOUNDI = "../foundico/foundico_enriched_from_joined.csv"
foundi.to_csv(OUT_FOUNDI, index=False)
print(f"‚úÖ Guardado: {OUT_FOUNDI}  ({len(foundi)} filas, {foundi.shape[1]} columnas)")

# Vista r√°pida de algunos duplicados actualizados
display(foundi.loc[mask_dup, ["symbol_std", "ico_successful", "soft_cap", "hard_cap", "amount_raised", "hit_soft_cap", "hit_hard_cap"]])

# Verificaci√≥n r√°pida de distribuci√≥n
print(f'Distribucion de ico_successful: {foundi["ico_successful"].value_counts(dropna=False)}')
```

# Detectar duplicados por s√≠mbolo dentro del dataset de FOUNDICO

```python
dupes = foundi[foundi.duplicated("symbol_std", keep=False)].sort_values("symbol_std")
print(f"üîé Duplicados por symbol_std: {dupes['symbol_std'].nunique()} s√≠mbolos con duplicados.")
display(dupes[["symbol_std","name_std","type","accepting","amount_raised","soft_cap","hard_cap"]].head(10))

def consolidate_duplicates(df):
    def combine_str(series):
        vals = (
            series.dropna()
            .astype(str)
            .apply(lambda s: s.strip())
            .unique()
        )
        return ", ".join(sorted(v for v in vals if v and v.lower() != "nan"))

    agg_rules = {
        "category": combine_str,
        "type": combine_str,
        "accepting": combine_str,
        "markets": combine_str,
        "ico_score": "mean",
        "soft_cap": "max",
        "hard_cap": "max",
        "amount_raised": "max",
        "ico_successful": "max",
    }

    grouped = (
        df.groupby(["symbol_std"], dropna=False)
        .agg(agg_rules)
        .reset_index()
    )
    return grouped

foundi_clean = consolidate_duplicates(foundi)
print(f"Antes: {len(foundi)}  ‚Üí  Despu√©s: {len(foundi_clean)} registros √∫nicos.")

display(foundi_clean.head(10))
#display(foundi_clean)
print(f'Distribucion de ico_successful: {foundi["ico_successful"].value_counts(dropna=False)}')
print(f'Distribucion de ico_successful: {foundi_clean["ico_successful"].value_counts(dropna=False)}')
```

```python
import pandas as pd
import numpy as np

def verificar_exito_y_contar(foundi: pd.DataFrame):
    # Asegurar tipo num√©rico
    for col in ["soft_cap", "hard_cap", "amount_raised"]:
        if col in foundi.columns:
            foundi[col] = pd.to_numeric(foundi[col], errors="coerce")

    # Calcular exito seg√∫n reglas
    cond_soft = foundi["soft_cap"].notna() & (foundi["amount_raised"] >= foundi["soft_cap"])
    cond_hard = foundi["soft_cap"].isna() & foundi["hard_cap"].notna() & (foundi["amount_raised"] >= foundi["hard_cap"])

    foundi["pred_exito"] = np.where(cond_soft | cond_hard, 1, 0)

    # Asegurar columna ico_successful como num√©rica / booleana
    if "ico_successful" in foundi.columns:
        foundi["ico_successful"] = foundi["ico_successful"].replace(
            {"True": 1, "true": 1, "Yes": 1, "yes": 1, "False": 0, "false": 0, "No": 0, "no": 0}
        ).astype("float")
    else:
        foundi["ico_successful"] = np.nan

    # Contadores
    same = ((foundi["ico_successful"] == foundi["pred_exito"]) & foundi["ico_successful"].notna()).sum()
    diff = ((foundi["ico_successful"] != foundi["pred_exito"]) & foundi["ico_successful"].notna()).sum()
    missing = (foundi["soft_cap"].notna() & foundi["amount_raised"].notna() & foundi["ico_successful"].isna()).sum()
    hitsoft = (foundi["hit_soft_cap"].notna() & foundi["ico_successful"].isna()).sum()

    print("üìä Comparaci√≥n pred_exito vs ico_successful:")
    print(f"  ‚úÖ Coinciden: {same}")
    print(f"  ‚ö†Ô∏è Difieren: {diff}")
    print(f"  ‚ùå ico_successful ausente: {missing}")
    print(f"  ‚ùå hitsoft: {hitsoft}")
    display(foundi[((foundi["ico_successful"] != foundi["pred_exito"]) & foundi["ico_successful"].notna())].loc[:,['amount_raised','soft_cap','hard_cap','symbol_std','ico_score','ico_successful','hit_soft_cap','hit_hard_cap','pred_exito']])
    #display(foundi[(foundi["hit_soft_cap"].notna() | foundi["hit_hard_cap"].notna()) & foundi["ico_successful"].isna()].loc[:,['amount_raised','soft_cap','hard_cap','symbol_std','ico_successful','hit_soft_cap','hit_hard_cap','pred_exito']])
    #display(foundi[(foundi["ico_successful"] != foundi["pred_exito"]) & foundi["ico_successful"].notna()].loc[:,['amount_raised','soft_cap','hard_cap','symbol_std','ico_successful','hit_soft_cap','hit_hard_cap','pred_exito']])
    #display(foundi[foundi["hard_cap"].notna() & foundi["amount_raised"].notna() & foundi["ico_successful"].isna()].loc[:,['amount_raised','soft_cap','hard_cap','symbol_std','ico_successful','hit_soft_cap','hit_hard_cap','pred_exito']])
    display(foundi[foundi["markets"].notna() & foundi["ico_successful"].isna()].loc[:,['amount_raised','soft_cap','hard_cap','symbol_std','ico_score','ico_successful','hit_soft_cap','hit_hard_cap','pred_exito']])
    return foundi

# Ejemplo de uso
foundi = verificar_exito_y_contar(foundi)
print(sorted(foundi.columns.tolist()))
print(sorted(joined.columns.tolist()))

# --- Resumen r√°pido ---
new_cols = [f"has_{c}" for c in cols_links]
print("‚úÖ Columnas creadas:", new_cols)
display(foundi[new_cols].head(10))

# Por curiosidad, mostrar cu√°ntos proyectos tienen cada red
print("\nüìä Porcentaje de ICOs con cada presencia:")
display((foundi[new_cols].mean() * 100).round(1).astype(str) + "%")
```

```python
from difflib import SequenceMatcher
import numpy as np

# Subconjuntos duplicados e index por symbol_std
joined_dup = joined[joined["symbol_std"].isin(duplis_list)].set_index("symbol_std")
foundi_dup = foundi[foundi["symbol_std"].isin(duplis_list)].set_index("symbol_std")

# Join con sufijos
merged_dup = joined_dup.join(foundi_dup, lsuffix="_ex", rsuffix="_fo", how="outer")

# === Mapeo manual de columnas equivalentes ===
MAPPING_HINTS = {
    # joined / foundico
    "accepts": "accepting",
    "amount_raised": "amount_raised",
    "hard_cap": "hard_cap",
    "has_github": "has_github",
    "has_reddit": "has_reddit",
    "hit_hard_cap": "hit_soft_cap",
    "name_std": "name_std",
    "symbol_std": "symbol_std",
    "ico_successful": "ico_successful",
    "whitelist": "whitelist",
    "website_available": "has_website",    # en FoundICO es URL ‚Üí pod√©s derivar flag despu√©s
    "whitepaper_available": "has_whitepaper",
    "kyc": "kyc",
    "jurisdiction": "location",
    "ico_start_date": "start_datetime",
    "ico_end_date": "end_datetime",
    "token_price_usd": "token_price",
    "token_type": "token_type",
    "tokens_for_sale": "tokens_for_sale",
    "min_investment_usd": "min_purchase",
    "max_investment_usd": "max_purchase",
    "role_of_token": "type",
    "rating": "ico_score",
}

#['', '', 'discount_max_pct', '', '', '', 'has_telegram', '', '', '', 'ico_length_actual_days', 'ico_length_planned_days', '', '', 'interest', 'is_ieo', 'is_tax_regulated', '', '', 'max_investment_usd', 'min_investment_usd', '', 'rating', 'roadmap_available', 'role_of_token', '', '', 'team_size', '', '', '', 'total_tokens', '', '', '']
#['', '', 'caps_unit', 'category', '', 'facebook', 'foundico_id', 'github', '', 'has_facebook', '', '', 'has_twitter', '', '', '', '', 'ico_score', '', 'ico_successful_rule', '', '', 'markets', 'max_purchase', 'min_purchase', 'mvp', '', 'platform', 'pred_exito', 'purchase_unit', 'reddit', '', '', '', '', 'token_price_unit', '', '', 'twitter', 'type', 'website', '', 'whitepaper']

# === Sugeridor por similitud para columnas no mapeadas expl√≠citas ===
def most_similar(col, candidates):
    sims = [(SequenceMatcher(None, col, c).ratio(), c) for c in candidates]
    sims.sort(reverse=True)
    return sims[0] if sims else (0, None)

# Construimos pares comparables (joined_col, foundi_col)
pairs = []
foundi_cols = set(foundi_dup.columns)
for joined_col in joined_dup.columns:
    # 1) si hay mapeo manual
    if joined_col in MAPPING_HINTS and MAPPING_HINTS[joined_col] in foundi_dup.columns:
        pairs.append((joined_col, MAPPING_HINTS[joined_col]))
        continue
    # 2) si existe igual nombre en found
    if joined_col in foundi_cols:
        pairs.append((joined_col, joined_col))
        continue
    # 3) fallback: similitud
    score, best = most_similar(joined_col, list(foundi_cols))
    if score >= 0.80 and best:
        pairs.append((joined_col, best))

# Funciones de normalizaci√≥n para comparaci√≥n
def norm_numeric(s):
    return pd.to_numeric(s, errors="coerce")

def norm_text(s):
    return s.astype(str).str.strip().str.lower().replace({"nan": np.nan, "": np.nan})

# Detectar contradicciones
contradicciones = []
for joined_col, foundi_col in pairs:
    joined_name = joined_col# + "_joined"
    foundi_name = foundi_col# + "_foundi"
    if joined_name not in merged_dup.columns or foundi_name not in merged_dup.columns:
        continue
    a = merged_dup[joined_name]
    b = merged_dup[foundi_name]

    # decidir si comparar como num√©rico o texto
    a_num, b_num = norm_numeric(a), norm_numeric(b)
    if a_num.notna().any() or b_num.notna().any():
        ax, bx = a_num, b_num
    else:
        ax, bx = norm_text(a), norm_text(b)

    mask = ax.notna() & bx.notna() & (ax != bx)
    if mask.any():
        tmp = merged_dup.loc[mask, [joined_name, foundi_name]].copy()
        tmp.insert(0, "feature_joined", joined_col)
        tmp.insert(1, "feature_foundi", foundi_col)
        contradicciones.append(tmp)

print(f"üìä Columnas comparadas: {len(pairs)}")
if contradicciones:
    diffs_df = pd.concat(contradicciones, axis=0)
    print(f"‚ö†Ô∏è Contradicciones encontradas: {len(diffs_df)} filas")
    display(diffs_df.head(10))
else:
    print("‚úÖ Sin contradicciones en las columnas comparadas.")
```

```python
import pandas as pd
import numpy as np

# ------------------ Config ------------------
FOUNDI_PATH = "../foundico/foundico_enriched_from_joined.csv"
JOINED_PATH = "../join/ico_union_canonical_v6.csv"

OUT_FULL = "../final/ico_dataset_final_v2.csv"
OUT_PREICO = "../final/ico_dataset_preico_features_v2.csv"

# ------------------ Helpers ------------------
def norm_symbol(s):
    return pd.Series(s, dtype="string").str.strip().str.upper()

def norm_name(s):
    return pd.Series(s, dtype="string").str.strip().str.lower()

def to_int64_safe(series):
    """Convierte a Int64 (nullable) sin reventar por NaN."""
    return pd.to_numeric(series, errors="coerce").astype("Int64")
    
def to_bool_safe(series):
    """Convierte a Int64 (nullable) sin reventar por NaN."""
    return pd.to_numeric(series, errors="coerce").astype("int")

def combine_str_unique(series):
    vals = (
        series.dropna()
        .astype(str)
        .str.strip()
        .replace({"", np.nan})
        .dropna()
        .unique()
    )
    return ", ".join(sorted(vals))

def ensure_has_flags(df):
    # crea has_* si no est√°n; si existen, los normaliza a Int64 (nullable)
    link_cols = ["website","whitepaper","github","twitter","facebook","reddit"]
    for col in link_cols:
        flag = f"has_{col}"
        if flag not in df.columns:
            if col in df.columns:
                df[flag] = (
                    pd.Series(df[col], dtype="string")
                    .str.strip().replace({"", np.nan})
                    .notna()
                    .astype("Int64")
                )
            else:
                df[flag] = pd.Series(pd.NA, index=df.index, dtype="Int64")
        else:
            # normalizar a Int64 respetando NaN
            df[flag] = to_int64_safe(df[flag])
            
    # ademas normalizamos:
    norm_cols = ['kyc', 'mvp', 'team_size', 'roadmap_available', 'is_ieo', 'is_tax_regulated', 'hit_softcap', 'hit_hardcap']
    # for col in norm_cols:
    #     # normalizar a Int64 respetando NaN
    #     df['has_'+col] = (
    #         pd.Series(df[col], dtype="string")
    #         .str.strip().replace({"", np.nan})
    #         .notna()
    #         .astype("Int64")
    #     )
    
    return df

# ------------------ Load ------------------
foundi = pd.read_csv(FOUNDI_PATH)
joined = pd.read_csv(JOINED_PATH)

# Normalizar claves
for df in (foundi, joined):
    if "symbol_std" in df.columns:
        df["symbol_std"] = norm_symbol(df["symbol_std"])
    else:
        df["symbol_std"] = pd.Series(pd.NA, index=df.index, dtype="string")

    if "name_std" in df.columns:
        df["name_std"] = norm_name(df["name_std"])
    else:
        df["name_std"] = pd.Series(pd.NA, index=df.index, dtype="string")

# ------------------ Mapeo de atributos ------------------
MAPPING_HINTS = {
    # joined -> foundico
    "accepts": "accepting",
    "amount_raised": "amount_raised",
    "hard_cap": "hard_cap",
    "has_github": "has_github",
    "has_reddit": "has_reddit",
    "hit_hard_cap": "hit_hard_cap",
    "hit_soft_cap": "hit_soft_cap",
    "name_std": "name_std",
    "symbol_std": "symbol_std",
    "ico_successful": "ico_successful",
    "whitelist": "whitelist",
    "website_available": "has_website",
    "whitepaper_available": "has_whitepaper",
    "kyc": "kyc",
    "jurisdiction": "location",
    "ico_start_date": "start_datetime",
    "ico_end_date": "end_datetime",
    "token_price_usd": "token_price",
    "token_type": "token_type",
    "tokens_for_sale": "tokens_for_sale",
    "min_investment_usd": "min_purchase",
    "max_investment_usd": "max_purchase",
    "role_of_token": "type",
    "industry": "category",
    "platform": "platform",
    "rating": "ico_score",
}

# Renombrar columnas del JOINED a nombres de FoundICO donde apliquen
joined_ren = joined.rename(columns=MAPPING_HINTS)


# ------------------ Merge externo por symbol_std ------------------
merged = pd.merge(
    foundi, joined_ren,
    on=["symbol_std"], how="outer",
    suffixes=("_fo", "_jo"), indicator=True
)


# ------------------ Resolver conflictos: prioriza FoundICO ------------------
all_cols = sorted(set(foundi.columns) | set(joined_ren.columns))
conflicts_log = []

for col in all_cols:
    cfo = f"{col}_fo" if f"{col}_fo" in merged.columns else None
    cjo = f"{col}_jo" if f"{col}_jo" in merged.columns else None

    if cfo and cjo:
        # preferimos FoundICO; si NaN, tomamos JOINED
        merged[col] = merged[cfo].where(merged[cfo].notna(), merged[cjo])
        # log de conflictos
        mask_conf = merged[cfo].notna() & merged[cjo].notna() & (merged[cfo].astype(str) != merged[cjo].astype(str))
        if mask_conf.any():
            conflicts_log.append((col, int(mask_conf.sum())))
    elif cfo:
        merged[col] = merged[cfo]
    elif cjo:
        merged[col] = merged[cjo]

# limpiar columnas auxiliares
drop_aux = [c for c in merged.columns if c.endswith("_fo") or c.endswith("_jo") or c == "_merge"]
final_full = merged.drop(columns=drop_aux)


# ------------------ Normalizar flags has_* y otros binarios ------------------
final_full = ensure_has_flags(final_full)


# mantener ico_successful como objetivo (no tocarlo ac√°), pero evitar castear a int duro
if "ico_successful" in final_full.columns:
    final_full["ico_successful"] = to_int64_safe(final_full["ico_successful"])


# ------------------ Guardar dataset maestro ------------------
final_full.to_csv(OUT_FULL, index=False)
print(f"‚úÖ Dataset maestro unido guardado: {OUT_FULL}  ({final_full.shape[0]} filas, {final_full.shape[1]} cols)")

if conflicts_log:
    print("‚ö†Ô∏è Conflictos (columna, casos):")
    for c, n in conflicts_log[:20]:
        print(f"  ‚Ä¢ {c}: {n}")
    if len(conflicts_log) > 20:
        print(f"  ‚Ä¶ +{len(conflicts_log)-20} columnas con conflictos")
```

### Feature engineering pre-ICO 
##### se agregan caracteristicas: temporales + ratios + accepts binarizados + social binarizados ) ===

```python
# === Feature engineering pre-ICO (temporal + logs + ratios + accepts binarizado) ===
import pandas as pd
import numpy as np
import re

# df debe existir: columnas esperadas (al menos algunas): 
# ['accepting','soft_cap','hard_cap','token_price','tokens_for_sale','min_purchase','max_purchase',
#  'category','platform','location','kyc','whitelist','mvp','has_website','has_whitepaper',
#  'has_github','has_twitter','has_facebook','has_reddit','team_size','roadmap_available',
#  'is_tax_regulated','start_datetime','ico_successful']

OUTPUT_FEATURE_ENGINEERING = "../final/ico_dataset_final_v2_clean_enriquecido_feature_engineering_v1.csv"
#df = preico.copy()
#df = df_ico_clean_enriq.copy()

### CARGA DEL DATASET ENRIQUECIDO CON CoinGecko y CMC ### 
IN_PATH = "../final/ico_dataset_final_v2_clean_enriquecido.csv"   # entrada de la celda anterior pero enriquecida con coingecko y coinmarketcap
df = pd.read_csv(IN_PATH)
assert "ico_successful" in df.columns, "Falta la etiqueta objetivo 'ico_successful'."
columnas_para_eliminar = []


### --- (1) Encodings Temporales (sin leakage) --- ###
def parse_dates_robust(s: pd.Series) -> pd.Series:
    s = s.astype(str).str.strip()
    # intentos con formatos comunes (para silenciar warnings)
    parsed = pd.to_datetime(s, format="%Y-%m-%d %H:%M:%S", errors="coerce")
    for fmt in ("%Y-%m-%d", "%d/%m/%Y", "%Y/%m/%d", "%d %b %Y", "%b %d, %Y"):
        need = parsed.isna()
        if need.any():
            parsed.loc[need] = pd.to_datetime(s[need], format=fmt, errors="coerce", dayfirst=True)
    # √∫ltimo fallback
    need = parsed.isna()
    if need.any():
        parsed.loc[need] = pd.to_datetime(s[need], errors="coerce", dayfirst=True)
    return parsed

if "start_datetime" in df.columns:
    dt = parse_dates_robust(df["start_datetime"])
else:
    dt = pd.Series(pd.NaT, index=df.index)

### Se crean las columnas start_year y start_qrt, para luego eliminar start_datetime ###
print("\n### --- (1) Encodings Temporales (sin leakage) --- ###")
print(f"-> Existe start_year: {'start_year' in df.columns}. Existe start_qtr: {'start_qtr' in df.columns}.")
df["start_year"]  = dt.dt.year
df["start_qtr"]   = dt.dt.quarter
print(f"-> Vacios en start_year: {df['start_year'].isna().sum()}. Vacios en start_qtr: {df['start_qtr'].isna().sum()}.")
columnas_para_eliminar.append("start_datetime")



### --- (2) Sincronizacion de los campos de redes sociales con su correspondiente has_{red} --- ###
redes = ["twitter", "facebook", "reddit", "github", "telegram", "whitepaper"]

def unificar_columnas_redes(df, redes=None):
    """
    Unifica columnas duplicadas de redes sociales (por ejemplo 'twitter' y 'twitter_url').
    Si ambas existen, conserva el valor no nulo.
    Si solo existe la versi√≥n con '_url', crea la versi√≥n sin '_url'.
    Finalmente, elimina las columnas *_url.
    """
    for red in redes:
        col_normal = red
        col_url = f"{red}_url"

        # Caso 1: existen ambas ‚Üí combinamos valores (prioriza no nulos)
        if col_normal in df.columns and col_url in df.columns:
            valor_no_nulo = df[col_normal].combine_first(df[col_url])
            print(f"-> Existen ambas para {col_normal}, se graba el valor priorizando no nulo.")
            df[col_normal] = valor_no_nulo

        # Caso 2: solo existe la versi√≥n con _url ‚Üí la renombramos
        elif col_url in df.columns and col_normal not in df.columns:
            print(f"-> Solo existe _url para {col_normal}, se crea la columna.")
            df[col_normal] = df[col_url]


    # Eliminamos todas las *_url que ya no se necesitan
    url_cols = [c for c in df.columns if c.endswith("_url")]
    df = df.drop(columns=url_cols, errors="ignore")
    return df

def sincronizar_flags_con_links(df, redes=None):
    """
    Si existe una columna 'xxx_link' y una 'has_xxx',
    pone has_xxx = 1 cuando xxx_link tiene valor (no nulo ni vac√≠o).
    """
    for red in redes:
        link_col = f"{red}"
        flag_col = f"has_{red}"
        
        if link_col in df.columns and flag_col in df.columns:
            print(f"{link_col} - {flag_col}")
            # Aseguramos el tipo int (0/1)
            df[flag_col] = (
                df[link_col]
                .apply(lambda x: 1 if isinstance(x, str) and x.strip() != "" else 0)
                .combine_first(df[flag_col])  # conserva valores previos cuando no hay link
                .fillna(0)
                .astype(int)
            )
    return df

def check_flags_de_redes_sincronizadas(redes):
    """
    Ver cu√°ntos casos hay por red donde existe link y si el has_{red} coincide.
    """
    for red in redes:
        link_col = f"{red}"
        flag_col = f"has_{red}"
        if link_col in df.columns and flag_col in df.columns:
            print(f"=== {red.upper()} ===")
            print(
                df.loc[df[link_col].notna(), flag_col]
                .value_counts(dropna=False)
                .rename_axis(flag_col)
                .reset_index(name="count")
            )
            
print("\n### --- (2) Sincronizacion de los campos de redes sociales con su correspondiente has_{red} --- ###")
df = unificar_columnas_redes(df, redes)
#check_flags_de_redes_sincronizadas(redes)
df = sincronizar_flags_con_links(df, redes)
#check_flags_de_redes_sincronizadas(redes)
for red in redes:
    columnas_para_eliminar.append(f"{red}_url")


### --- 3) Transformaciones log1p para estabilizar escalas (solo pre-ICO) --- ###
LOG_NUMS = [
    "soft_cap", "hard_cap", "token_price",
    "tokens_for_sale", "min_purchase", "max_purchase",
    "team_size"
]
print("\n### --- 3) Transformaciones log1p para estabilizar escalas (solo pre-ICO) --- ###")
for col in LOG_NUMS:
    count_log1p = 0
    if col in df.columns:
        col_num = pd.to_numeric(df[col], errors="coerce")
        # log1p requiere valores >= -1; truncamos a 0 los negativos espurios
        col_num = col_num.mask(col_num < 0, np.nan)
        df[f"log1p_{col}"] = np.log1p(col_num)
        count_log1p += 1
print(f"-> Se estabilizaron {count_log1p} campos.")
columnas_para_eliminar.append(f"soft_cap_usd")
columnas_para_eliminar.append(f"hard_cap_usd")
columnas_para_eliminar.append(f"token_price_cap_usd")


    
### --- (4) Ratios/Interacciones (sin usar variables post-ICO) --- ###
print("\n### --- (4) Ratios/Interacciones (sin usar variables post-ICO) --- ###")
def safe_div(a, b):
    a = pd.to_numeric(a, errors="coerce")
    b = pd.to_numeric(b, errors="coerce")
    out = a / b
    out.replace([np.inf, -np.inf], np.nan, inplace=True)
    return out

# proporci√≥n objetivo alcanzado (sin usar 'amount_raised' para evitar leakage)
# Usamos solo relaciones entre objetivos y condiciones de venta
if "soft_cap" in df.columns and "hard_cap" in df.columns:
    df["cap_ratio_soft_over_hard"] = safe_div(df["soft_cap"], df["hard_cap"])

# Intensidad de min/max ticket (si ambos existen)
if "min_purchase" in df.columns and "max_purchase" in df.columns:
    df["ticket_minmax_ratio"] = safe_div(df["min_purchase"], df["max_purchase"])

# ‚Äúprecio relativo a la meta‚Äù (magnitud) ‚Äì √∫til aunque no causal directo
if "token_price" in df.columns and "soft_cap" in df.columns:
    df["price_to_softcap_ratio"] = safe_div(df["token_price"], df["soft_cap"])

# Densidad de oferta respecto al hard cap (si el token_price est√° definido)
# aprox: cu√°nto ‚Äúcab√≠a vender‚Äù al precio listado; si falta alguno, queda NaN
if all(c in df.columns for c in ["tokens_for_sale", "token_price", "hard_cap"]):
    denom = df["hard_cap"] / pd.to_numeric(df["token_price"], errors="coerce")
    df["sale_vs_capacity_ratio"] = safe_div(df["tokens_for_sale"], denom)


# --- 4) Binarizaci√≥n de 'accepting' ---
# Normaliza separadores (",", "/", " ", ";") y saca s√≠mbolos tipo "$" delante de tickers
# def split_accepts(x):
#     if pd.isna(x): 
#         return []
#     s = str(x).upper().replace("|", ",")
#     s = re.sub(r"[/; ]+", ",", s)           # separadores varios -> coma
#     s = re.sub(r"\$([A-Z0-9]+)", r"\1", s)  # quita prefijo $
#     s = s.replace("FIAT", "FIAT")
#     # normalizaciones simples
#     s = s.replace("USDT,", "USDT,").replace("USDC,", "USDC,")
#     parts = [p.strip() for p in s.split(",") if p.strip()]
#     return parts

# ACCEPTS_CANON = ["ETH","BTC","USDT","USDC","BNB","BUSD","SOL","MATIC","LTC","XRP","BCH","FIAT"]
# if "accepting" in df.columns:
#     accepts_series = df["accepting"].apply(split_accepts)
# else:
#     accepts_series = pd.Series([[]]*len(df), index=df.index)

#for sym in ACCEPTS_CANON:
#    df[f"accepts_{sym.lower()}"] = accepts_series.apply(lambda lst: int(sym in lst) if isinstance(lst, list) else 0)


### --- (5) Binarizaci√≥n de accepting -> solo dos flags: accepts_BTC y accepts_ETH, los demas no son relevantes --- ###
print("\n### --- (5) Binarizaci√≥n de accepting -> solo dos flags: accepts_BTC y accepts_ETH, los demas no son relevantes --- ###")
def add_accepts_flags(df, col="accepting"):
    acc = df[col].fillna("").str.upper()
    df["accepts_BTC"] = acc.str.contains(r"\bBTC\b|BITCOIN", regex=True).astype(int)
    df["accepts_ETH"] = acc.str.contains(r"\bETH\b|ETHEREUM", regex=True).astype(int)
    return df

df = add_accepts_flags(df, col="accepting")
print(f"-> Vacios en accepts_BTC: {df['accepts_BTC'].isna().sum()}. Vacios en accepts_ETH: {df['accepts_ETH'].isna().sum()}.")
columnas_para_eliminar.append(f"accepting")



### --- (6) Reducir cardinalidad de 'category' y 'location' --- ###
print("\n### --- (6) Reducir cardinalidad de 'category', 'platform' y 'location' --- ###")
#    - Nos quedamos con Top-K por frecuencia y todo lo dem√°s va a "Other"
def cap_high_cardinality(df, col, top_k=15, other_label="Other"):
    top_vals = df[col].value_counts(dropna=True).head(top_k).index
    # Reemplazo: todo lo que no est√© en top_vals pasa a Other (NaN tambi√©n)
    df[col] = np.where(df[col].isin(top_vals), df[col], other_label)
    return df

category_agrupado = df['category'].value_counts()
location_agrupado = df['location'].value_counts()
platform_agrupado = df['platform'].value_counts()
print(f"-> Antes: category_agrupado en 'Other': {category_agrupado.get('Other', 0)}")
print(f"-> Antes: location_agrupado en 'Other': {location_agrupado.get('Other', 0)}")
print(f"-> Antes: platform_agrupado en 'Other': {platform_agrupado.get('Other', 0)}")

df = cap_high_cardinality(df, "category", top_k=15, other_label="Other")
df = cap_high_cardinality(df, "location", top_k=15, other_label="Other")
df = cap_high_cardinality(df, "platform", top_k=10, other_label="Other")

category_agrupado = df['category'].value_counts()
location_agrupado = df['location'].value_counts()
platform_agrupado = df['platform'].value_counts()
print(f"-> Despues: category_agrupado en 'Other': {category_agrupado.get('Other', 0)}")
print(f"-> Despues: location_agrupado en 'Other': {location_agrupado.get('Other', 0)}")
print(f"-> Despues: platform_agrupado en 'Other': {platform_agrupado.get('Other', 0)}")



### --- (7) Unificacion de nombres de paises --- ###
print("\n### --- (7) Unificacion de nombres de paises --- ###")

location_agrupado = df['location'].value_counts()
print(f"-> Antes: United States of America: {location_agrupado.get('United States of America', 0)}")
print(f"-> Antes: United States: {location_agrupado.get('United States', 0)}")
print(f"-> Antes: USA: {location_agrupado.get('USA', 0)}")

df['location'] = df['location'].replace("United States of America", "USA")
df['location'] = df['location'].replace("United States", "USA")

location_agrupado = df['location'].value_counts()
print(f"-> Despues: United States of America: {location_agrupado.get('United States of America', 0)}")
print(f"-> Despues: United States: {location_agrupado.get('United States', 0)}")
print(f"-> Despues: USA: {location_agrupado.get('USA', 0)}")



### --- (8) Saneos r√°pidos de flags/categ√≥ricas num√©ricas (0/1/NaN) --- ###
print("\n### --- (8) Saneos r√°pidos de flags/categ√≥ricas num√©ricas (0/1/NaN) --- ###")
BIN_COLS = ["kyc","whitelist","mvp","has_website","has_whitepaper","has_github", "has_telegram",
            "has_twitter","has_facebook","has_reddit","roadmap_available","is_tax_regulated"]
for c in BIN_COLS:
    if c in df.columns:
        s = pd.to_numeric(df[c], errors="coerce")
        # fuerza al dominio {0,1} si ven√≠an como 0.0/1.0 o strings; deja NaN si no matchea
        s = s.where(s.isin([0,1]), np.nan)
        df[c] = s.astype("Int64")



### --- (9) Imputar 0 en binarios sociales si faltan (ausencia = 0) --- ###
print("\n### --- (9) Imputar 0 en binarios sociales si faltan (ausencia = 0) --- ###")
binary_like = [
    "kyc","whitelist","mvp","has_website","has_whitepaper","has_github",
    "has_twitter","has_facebook","has_reddit","is_tax_regulated"
]
for col in binary_like:
    if col in df.columns:
        df[col] = df[col].fillna(0).astype(int)



### --- (10) Reconstruir ticket_minmax_ratio cuando falte y haya datos --- ###
print("\n### --- (10) Reconstruir ticket_minmax_ratio cuando falte y haya datos --- ###")
if "ticket_minmax_ratio" in df.columns and "min_purchase" in df.columns and "max_purchase" in df.columns:
    mask_can_build = df["ticket_minmax_ratio"].isna() & df["min_purchase"].notna() & df["max_purchase"].notna() & (df["max_purchase"] != 0)
    df.loc[mask_can_build, "ticket_minmax_ratio"] = df.loc[mask_can_build, "min_purchase"] / df.loc[mask_can_build, "max_purchase"]



### --- (11) Se graba el dataset final (NO pre-ico) para despues obtener el pre-ico definitivo --- ###
df.to_csv(OUTPUT_FEATURE_ENGINEERING, index=False)
print(f'\n### --- (11) Se graba el dataset final (NO pre-ico) en: {OUTPUT_FEATURE_ENGINEERING} --- ###')



### --- (12) Se imputan valor del target cuando sea posible --- ###
print("\n### --- (12) Se imputan valor del target cuando sea posible --- ###")
print(f'Distribucion de ico_successful: {df["ico_successful"].value_counts(dropna=False)}')
imputar_valores_ico_successful(df, 3.9, 5.9)
print(f'Distribucion de ico_successful: {df["ico_successful"].value_counts(dropna=False)}')



### --- (13) Crear flag si hay valor en contract_address --- ###
print("\n### --- (13) Crear flag si hay valor en contract_address --- ###")
if "contract_address" in df.columns:
    df["has_contract_address"] = (
        df["contract_address"]
        .apply(lambda x: 1 if isinstance(x, str) and x.strip() != "" else 0)
        .astype(int)
    )
    columnas_para_eliminar.append(f"contract_address")



### --- (14) Mini-reporte ---
print("\n### --- (14) Mini-reporte --- ###")
added_cols = [c for c in df.columns if c.startswith("log1p_") or 
              c.endswith("_ratio") or c.startswith("accepts_") or c in ["start_year","start_qtr","start_month","start_dow"]]
print(f"‚úÖ Feature engineering aplicado. Columnas nuevas: {len(added_cols)}")
print(sorted(added_cols)[:20], " ...")
print("\nEjemplo (primeras 3 filas, columnas clave):")
show_cols = ["start_year","start_qtr"
             "log1p_soft_cap","log1p_hard_cap","log1p_token_price","log1p_tokens_for_sale",
             "cap_ratio_soft_over_hard","ticket_minmax_ratio","price_to_softcap_ratio","sale_vs_capacity_ratio",
             "accepts_eth","accepts_btc","accepts_usdt","accepts_usdc","accepts_bnb","accepts_fiat"]
show_cols = [c for c in show_cols if c in df.columns]
display(df[show_cols].head(3))
```

```python
def columnas_dominantes(df, umbral=0.9):
    """
    Muestra las columnas num√©ricas o booleanas en las que
    un √∫nico valor domina en m√°s del 'umbral' (por defecto 90%).
    Devuelve un DataFrame con la proporci√≥n del valor m√°s frecuente.
    """
    cols = df.select_dtypes(include=["number", "bool"]).columns
    dominantes = {}

    for col in cols:
        if df[col].notna().sum() == 0:
            continue  # saltamos columnas totalmente vac√≠as
        frec_max = df[col].value_counts(normalize=True, dropna=False).iloc[0]
        if frec_max >= umbral:
            dominantes[col] = round(frec_max, 3)

    if dominantes:
        resumen = pd.DataFrame(list(dominantes.items()), columns=["columna", "proporcion_valor_dominante"])
        print(f"Columnas con m√°s del {umbral*100:.0f}% de un mismo valor: {resumen}")
        return resumen.sort_values("proporcion_valor_dominante", ascending=False)
    else:
        print(f"No hay columnas con m√°s del {umbral*100:.0f}% de un mismo valor.")
        return pd.DataFrame(columns=["columna", "proporcion_valor_dominante"])
```

##### Incluye: selecci√≥n din√°mica de columnas, imputaci√≥n (num√©ricas/booleanas/categ√≥ricas), one-hot encoding con handle_unknown='ignore', scaling para num√©ricas, y generaci√≥n de los nombres de features transformados.

```python
%%time

# Preprocesamiento listo para modelar (OneHot + Imputers + Scaling)
import pandas as pd
import numpy as np
import joblib
import os
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.utils import Bunch

# ---------------------------
# Config
# ---------------------------
ARTIFACTS_DIR = "../artifacts"
INPUT_DATASET_FEATURE_ENG = "../final/ico_dataset_preico_feature_engineering_v2.csv"  # <-- el csv de la celda anterior con feature engineering
OUTPUT_PREPROCECED_BUNDLE = f"{ARTIFACTS_DIR}/ico_preproc_bundle_v2.pk1"  # <-- este bundle tendra todo los datos necesarios para entrenar los modelos
if not os.path.exists(ARTIFACTS_DIR):
    os.mkdir(ARTIFACTS_DIR)

# Variable objetivo
TARGET   = "ico_successful"

# Categ√≥ricas esperadas (si no existen, se ignoran)
CATEGORICAL_CANDIDATES = ["category"] # ["category", "platform", "location"]

# Prefijos/he√∫r√≠stica para columnas binarias
BINARY_PREFIXES = ("has_", "accepts_")
BINARY_EXTRAS   = {"kyc", "whitelist", "mvp", "roadmap_available", "is_tax_regulated"}

# ---------------------------
# Carga y selecci√≥n de columnas
# ---------------------------
df_feat_eng = pd.read_csv(INPUT_DATASET_FEATURE_ENG)

# Asegurar que el target existe y sin NaN
if TARGET not in df_feat_eng.columns:
    raise ValueError(f"'{TARGET}' no est√° en el CSV.")
df_feat_eng = df_feat_eng[df_feat_eng[TARGET].notna()].copy()

# Forzar target a 0/1 (por si llega como float/str)
df_feat_eng[TARGET] = df_feat_eng[TARGET].astype(float).round().astype(int)

# Detectar categ√≥ricas (presentes en el dataset)
categorical_cols = [c for c in CATEGORICAL_CANDIDATES if c in df_feat_eng.columns]

# Detectar binarias por nombre y extras
binary_cols = []
for c in df_feat_eng.columns:
    if c == TARGET: 
        continue
    if c.startswith(BINARY_PREFIXES) or c in BINARY_EXTRAS:
        if c in df_feat_eng.columns:
            binary_cols.append(c)
binary_cols = [c for c in binary_cols if c in df_feat_eng.columns]

# Num√©ricas = todas las num√©ricas excluyendo el target y excluyendo las binarias (las tratamos aparte)
numeric_cols_all = df_feat_eng.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols = [c for c in numeric_cols_all if c not in set(binary_cols) | {TARGET}]

# Algunas columnas num√©ricas pueden estar como 'object' (ej. token_price con texto raro)
# for c in df_feat_eng.columns:
#     if c not in categorical_cols and c != TARGET and df_feat_eng[c].dtype == "object":
#         df_feat_eng[c] = pd.to_numeric(df_feat_eng[c], errors="coerce")

# ---------------------------
# Train/Test split (estratificado por clase)
# ---------------------------
X = df_feat_eng.drop(columns=[TARGET])
y = df_feat_eng[TARGET].astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# ---------------------------
# Transformadores por bloque
# ---------------------------
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler",  StandardScaler(with_mean=True, with_std=True)),
])

binary_transformer = Pipeline(steps=[
    # Para flags 0/1: si falta, imputamos 0 (ausencia) y NO escalamos
    ("imputer", SimpleImputer(strategy="constant", fill_value=0)),
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe",     OneHotEncoder(handle_unknown="ignore", sparse_output=False)),
])

# ---------------------------
# ColumnTransformer
# ---------------------------
transformers = []
if numeric_cols:
    transformers.append(("num",  numeric_transformer, numeric_cols))
if binary_cols:
    transformers.append(("bin",  binary_transformer,  binary_cols))
if categorical_cols:
    transformers.append(("cat",  categorical_transformer, categorical_cols))

preprocessor = ColumnTransformer(
    transformers=transformers,
    remainder="drop",  # descart√° columnas no listadas
    verbose_feature_names_out=True
)

# Ajustar en train y transformar
X_train_prep = preprocessor.fit_transform(X_train)
X_test_prep  = preprocessor.transform(X_test)

# ---------------------------
# Nombres de features transformados
# ---------------------------
feat_names = preprocessor.get_feature_names_out()
print(feat_names)

# Saneos finales
def _check_no_nans(a, name):
    if np.isnan(a).any() or np.isinf(a).any():
        raise ValueError(f"El conjunto {name} contiene NaN/Inf tras el preprocesamiento.")
_check_no_nans(X_train_prep, "X_train_prep")
_check_no_nans(X_test_prep,  "X_test_prep")

# Objeto con las cosas necesarias para los modelos
bundle = Bunch(
    X_train=X_train,
    X_test=X_test,
    y_train=y_train,
    y_test=y_test,
    X_train_prep=X_train_prep,
    X_test_prep=X_test_prep,
    feature_names=feat_names,
    preprocessor=preprocessor,
    categorical_cols=categorical_cols,
    numeric_cols=numeric_cols,
    binary_cols=binary_cols,
)

print("‚úÖ Preprocesamiento listo.")
print(f"- Categ√≥ricas: {bundle.categorical_cols}")
print(f"- Binarias:    {bundle.binary_cols[:10]}{' ...' if len(bundle.binary_cols)>10 else ''}")
print(f"- Num√©ricas:   {bundle.numeric_cols[:10]}{' ...' if len(bundle.numeric_cols)>10 else ''}")
print(f"- X_train_prep: {bundle.X_train_prep.shape}, X_test_prep: {bundle.X_test_prep.shape}")
print(f"- Features transformadas: {len(bundle.feature_names)}")

joblib.dump(bundle, OUTPUT_PREPROCECED_BUNDLE, compress=3)
print(f"‚úÖ Guardado: {OUTPUT_PREPROCECED_BUNDLE}")
```

```python
# =============================================
# üìä Celda 1 ‚Äî Gr√°ficos extendidos del dataset
# =============================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid", palette="viridis", font_scale=1.1)

df = pd.read_csv("../final/ico_dataset_preico_feature_engineering.csv")

# --- 1Ô∏è‚É£ Distribuci√≥n temporal de ICOs (ya la ten√≠as, pero ahora suavizada) ---
plt.figure(figsize=(8,4))
sns.lineplot(x=df['start_year'].value_counts().sort_index().index,
             y=df['start_year'].value_counts().sort_index().values,
             marker="o")
plt.title("Evoluci√≥n temporal de ICOs")
plt.xlabel("A√±o")
plt.ylabel("Cantidad de ICOs")
plt.show()

# --- 2Ô∏è‚É£ Proporci√≥n de ICOs exitosas por a√±o ---
plt.figure(figsize=(8,4))
df_year = df.groupby('start_year')['ico_successful'].mean().dropna()
sns.lineplot(x=df_year.index, y=df_year.values, marker="o", color="teal")
plt.title("Proporci√≥n de ICOs exitosas por a√±o")
plt.xlabel("A√±o de inicio")
plt.ylabel("Tasa de √©xito promedio")
plt.show()

# üåç Proporci√≥n de ICO exitosas por pa√≠s
min_icos = 40  # para evitar ruido de pa√≠ses con pocas ICOs
df_country = df.groupby("location").agg(
    total_icos=("ico_successful", "count"),
    success_rate=("ico_successful", "mean")
).query("total_icos >= @min_icos").sort_values("success_rate", ascending=False)

plt.figure(figsize=(9,5))
sns.barplot(x=df_country["success_rate"], y=df_country.index, palette="crest")
plt.title("Proporci√≥n de ICOs exitosas por pa√≠s (m√≠nimo 20 ICOs)")
plt.xlabel("Tasa de √©xito promedio")
plt.ylabel("Pa√≠s")
plt.xlim(0,1)
plt.show()

# üè≠ Proporci√≥n de ICO exitosas por industria
min_icos = 60
df_industry = df.groupby("category").agg(
    total_icos=("ico_successful", "count"),
    success_rate=("ico_successful", "mean")
).query("total_icos >= @min_icos").sort_values("success_rate", ascending=False)

plt.figure(figsize=(9,5))
sns.barplot(x=df_industry["success_rate"], y=df_industry.index, palette="mako")
plt.title("Proporci√≥n de ICOs exitosas por industria (m√≠nimo 20 ICOs)")
plt.xlabel("Tasa de √©xito promedio")
plt.ylabel("Industria / Categor√≠a")
plt.xlim(0,1)
plt.show()

# --- 3Ô∏è‚É£ Distribuci√≥n logar√≠tmica de variables financieras ---
plt.figure(figsize=(10,5))
for col in ['soft_cap','hard_cap','token_price']:
    sns.kdeplot(np.log1p(df[col].dropna()), label=col)
plt.title("Distribuci√≥n logar√≠tmica de variables financieras")
plt.xlabel("log(1+x)")
plt.legend()
plt.show()

# --- 4Ô∏è‚É£ Relaci√≥n entre hard_cap y √©xito ---
plt.figure(figsize=(6,5))
sns.boxplot(x='ico_successful', y=np.log1p(df['hard_cap']), data=df, palette="viridis")
plt.title("Hard Cap (log) seg√∫n resultado de la ICO")
plt.xlabel("ICO exitosa (1) / fallida (0)")
plt.ylabel("log(hard_cap)")
plt.show()

# --- 5Ô∏è‚É£ Distribuci√≥n de plataformas m√°s comunes ---
plt.figure(figsize=(9,5))
top_platforms = df['platform'].value_counts().nlargest(8)
sns.barplot(x=top_platforms.values, y=top_platforms.index)
plt.title("Plataformas m√°s utilizadas en ICOs")
plt.xlabel("Cantidad de proyectos")
plt.ylabel("Plataforma")
plt.show()

# --- 6Ô∏è‚É£ Conteo de ICOs por pa√≠s (location) ---
plt.figure(figsize=(9,6))
top_countries = df['location'].value_counts().nlargest(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 pa√≠ses de origen de ICOs")
plt.xlabel("Cantidad de ICOs")
plt.ylabel("Pa√≠s")
plt.show()

# --- 7Ô∏è‚É£ Mapa de correlaci√≥n enfocado en variables financieras ---
plt.figure(figsize=(8,6))
corr = df[['soft_cap','hard_cap','token_price','tokens_for_sale','min_purchase','max_purchase']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlaci√≥n entre variables financieras")
plt.show()

# --- 8Ô∏è‚É£ Pairplot entre variables financieras clave ---
sns.pairplot(df[['soft_cap','hard_cap','token_price','ico_successful']].dropna(),
             hue='ico_successful', corner=True, diag_kind='kde')
plt.suptitle("Relaciones entre variables financieras y √©xito", y=1.02)
plt.show()
```
